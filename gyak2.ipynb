# 2. feladat
## Írjuk fel a 5342634632512543646623001324 szám prímfelbontását!

```python
import sympy
factors = sympy.primefactors(5342634632512543646623001324)
factorint = sympy.factorint(5342634632512543646623001324)
mult = ""
for i in factors:
  mult += str(i) + "*" + str(factorint[i]) + "+"
print(mult)
```

# 4. feladat
## Bontsuk fel a zárójeleket a ((x+y+12)^2 + (x+y-11)^3)^10 kifejezésben!

```python
import sympy
x = sympy.Symbol('x')
y = sympy.Symbol('y')
print((x+y)**2)

polynomial = ((x+y+12)**2 + (x+y-11)**3)**10
print(polynomial.expand())
```

# 5. feladat
## Alakítsuk szorzattá a a^3b+b^3c+c^3a-ab^3-bc^3-ca^3 kifejezést!

```python
import sympy
a = sympy.Symbol('a')
b = sympy.Symbol('b')
c = sympy.Symbol('c')
polynomial = a**3*b+b**3*c+c**3*a-a*b**3-b*c**3-c*a**3
polynomial.factor()
```

# 7. feladat
## Készítsen egy olyan függvényt, amely kirajzol egy olyan gráfot, aminek csúcsai az összes egyjegyű pozitív szám és csak azon két csúcs között húzódik él, amelynek összege páros vagy hárommal osztható szám!

```python
import graphviz
def ex7():
  g = graphviz.Graph()
  for i in range(1,10):
    g.node(str(i))
    for j in range(1, i+1):
      if (i+j) % 2 == 0:
        g.edge(str(i), str(j), color='blue')
      elif (i+j) % 3 == 0:
        g.edge(str(i), str(j), color='green')
  return g
g1 = ex7()
g1
```

# 8. feladat
## Adjon meg egy függvényt, ami paraméterben kapja az X és Y szorzószámait és a létrehozandó polinom fokát, majd visszatér az új polinommal!

```python
import sympy
def Ex8(a,b,n):
  x = sympy.Symbol('x')
  y = sympy.Symbol('y')
  return (a*x + b*y)**n
print(Ex8(5,-4,3))
print(Ex8(5,-4,3).expand())
```
